# Reporting: plot price data

This module is the core for the production of the images that can be used for visually assessing the price data.

It is based on `CoinFigure` object that is a matplotlib plot optimized for price data plotting.

First, import the required packages.

```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
from surfingcrypto import Config,TS
from surfingcrypto.reporting.figures import CoinFigure
```

Then load the configuration files.

```python
parent="/Users/giorgiocaizzi/Documents/GitHub/surfingcrypto/"
c=Config(parent+"config")
```

Produce a comples `ta` plot for each coin specified in the configuration file.

```python
for coin in c.coins:
    ts=TS(c,coin=coin)
    fig=CoinFigure(
        ts,
        kind="ta",
        trendlines=False,
        coin=coin,
        graphstart="3m"
        )
```
